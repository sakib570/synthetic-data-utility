```r
# Read results
avila_result <- read.csv('..//data//results_csv_files//avila_correlation_data.csv')
avila_result
```

```r
# For testing without Sample synthesizer results
avila_result <- avila_result[-c(18),] 
avila_result
```

```r
# Calculate deviance from oriignal
library(tidyverse)

avila_result <- avila_result %>% mutate(Dev_Org_Auth_Clsf = abs(Author.Classifcation - Author.Classifcation[Method == "Original"]))
avila_result
```

```r
# Correlation author & CIO
author_cio <- cor.test(avila_result$Dev_Org_Auth_Clsf, avila_result$CIO.Regression.Fits.Raab..2016..m.10, 
                    method = "pearson")
author_cio
```

```r
# Correlation author & APO
author_apo <- cor.test(avila_result$Dev_Org_Auth_Clsf, avila_result$APO.above.90..Raab..2016..m.10, 
                    method = "pearson")
author_apo
```

```r
# Correlation author & KL-divergence
author_kl <- cor.test(avila_result$Dev_Org_Auth_Clsf, avila_result$Average.KL.Dv.over.all.variables.and.normalized.using.sample, 
                    method = "pearson")
author_kl
```

```r
# Correlation CIO & APO
cio_apo <- cor.test(avila_result$CIO.Regression.Fits.Raab..2016..m.10, avila_result$APO.above.90..Raab..2016..m.10, 
                    method = "pearson")
cio_apo
```

```r
# Correlation CIO & KL-divergence
cio_kl <- cor.test(avila_result$CIO.Regression.Fits.Raab..2016..m.10, avila_result$Average.KL.Dv.over.all.variables.and.normalized.using.sample, 
                    method = "pearson")
cio_kl
```

```r
# Correlation APO & KL-divergence
kl_apo <- cor.test(avila_result$Average.KL.Dv.over.all.variables.and.normalized.using.sample, avila_result$APO.above.90..Raab..2016..m.10, 
                    method = "pearson")
kl_apo
```

```r
# Read units results
avila_units_result <- read.csv('../data/results_csv_files/avila_author_class_units_counts.csv')
```

```r
# For testing without Sample synthesizer results
avila_units_result <- avila_units_result[-c(18),] 
avila_units_result
```

```r
# Calculate devitain from original
avila_units_result <- avila_units_result %>% mutate(Dev_Org_Class_A = abs(Class.A - Class.A[Method == "Original"]))
avila_units_result <- avila_units_result %>% mutate(Dev_Org_Class_F = abs(Class.F - Class.F[Method == "Original"]))
avila_units_result <- avila_units_result %>% mutate(Dev_Org_Class_H = abs(Class.H - Class.H[Method == "Original"]))
avila_units_result <- avila_units_result %>% mutate(Dev_Org_Class_E = abs(Class.E - Class.E[Method == "Original"]))
avila_units_result <- avila_units_result %>% mutate(Dev_Org_Class_I = abs(Class.I - Class.I[Method == "Original"]))
avila_units_result <- avila_units_result %>% mutate(Dev_Org_Class_Y = abs(Class.Y - Class.Y[Method == "Original"]))
avila_units_result <- avila_units_result %>% mutate(Dev_Org_Class_D = abs(Class.D - Class.D[Method == "Original"]))
avila_units_result <- avila_units_result %>% mutate(Dev_Org_Class_X = abs(Class.X - Class.X[Method == "Original"]))
avila_units_result <- avila_units_result %>% mutate(Dev_Org_Class_G = abs(Class.G - Class.G[Method == "Original"]))
avila_units_result <- avila_units_result %>% mutate(Dev_Org_Class_W = abs(Class.W - Class.W[Method == "Original"]))
avila_units_result <- avila_units_result %>% mutate(Dev_Org_Class_C = abs(Class.C - Class.C[Method == "Original"]))
avila_units_result <- avila_units_result %>% mutate(Dev_Org_Class_B = abs(Class.B - Class.B[Method == "Original"]))
avila_units_result
```

```r
# Correlation CIO & Units Class A
units_claa_a_cio <- cor.test(avila_units_result$Dev_Org_Class_D, avila_result$CIO.Regression.Fits.Raab..2016..m.10, 
                    method = "pearson")
units_claa_a_cio
```

```r
# Correlation APO & Units Class A
units_claa_a_apo <- cor.test(avila_units_result$Dev_Org_Class_A, avila_result$APO.above.90..Raab..2016..m.10, 
                    method = "pearson")
units_claa_a_apo
```

```r
# Correlation KL-divergence & Units Class A
units_claa_a_kl <- cor.test(avila_units_result$Dev_Org_Class_A, avila_result$Average.KL.Dv.over.all.variables.and.normalized.using.sample, 
                    method = "pearson")
units_claa_a_kl
```
