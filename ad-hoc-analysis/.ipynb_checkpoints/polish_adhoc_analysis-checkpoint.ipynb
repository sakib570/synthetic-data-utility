```r
vars <- c("sex", "age", "edu", "marital", "income",
          "ls", "trust", "nofriend", "smoke", "alcabuse", 
          "wkabint", "englang", "height", "weight")
```

```r
# prepare original dataset
SD2011M <- read.csv('../Ex-Jobb/SD2011M2.csv')
#summary.data.frame(polish)

polish <- SD2011M[, vars]

for(v in vars){
  if(is.factor(polish[,v])){
    polish[,v] <- as.character(polish[,v])
    polish[,v] <- factor(polish[,v],levels(SD2011M[,v]))
  }
}

polish$wkabint <- as.character(polish$wkabint)
polish$wkabint[polish$wkabint == "YES, TO EU COUNTRY" | 
              polish$wkabint == "YES, TO NON-EU COUNTRY"] <- "YES"
polish$wkabint <- factor(polish$wkabint,c("YES","NO"))

polish$marital <- as.character(polish$marital)
polish$marital[polish$marital == "LEGALLY SEPARATED" | 
              polish$marital == "DE FACTO SEPARATED"] <- "DIVORCED"
polish$marital <- factor(polish$marital,c("SINGLE","MARRIED","WIDOWED","DIVORCED"))

varsTest <- c("sex", "age", "height", "weight", "edu", "englang",
              "marital", "wkabint", "income", "ls", "trust",
              "nofriend", "smoke", "alcabuse")
```

```r
# prepare synthetic datasets
sds_polish_m <- paste0("sds","Polish",c("Par","Par","Dec","Dec"),"M",c("","T"))
sds_polish_cat_cac <- paste0("sds","PolishCatM",c("P","P","C","C"),"",c("","T"))
sds_polish_mo <- paste0("sds","Polish",c("Par","Par","Dec","Dec"),"MO",c("","T"))
sds_polish_mv <- paste0("sds","Polish",c("Par","Par","Dec","Dec"),"MV",c("","T"))
sds_polish_samp <- paste0("sds","PolishSamp",c(""),"",c(""))
sds_names <- c(sds_polish_m, sds_polish_cat_cac, sds_polish_mo, sds_polish_mv, sds_polish_samp)
sds_names <- str_sort(sds_names)
sds_names

sdsPolishParM <- eval(as.name(sds_polish_m[[1]]))
sdsPolishParMT <- eval(as.name(sds_polish_m[[2]]))
sdsPolishDecM <- eval(as.name(sds_polish_m[[3]]))
sdsPolishDecMT <- eval(as.name(sds_polish_m[[4]]))

sdsPolishCatMP <- eval(as.name(sds_polish_cat_cac[[1]]))
sdsPolishCatMPT <- eval(as.name(sds_polish_cat_cac[[2]]))
sdsPolishCatMC <- eval(as.name(sds_polish_cat_cac[[3]]))
sdsPolishCatMCT <- eval(as.name(sds_polish_cat_cac[[4]]))

sdsPolishParMO <- eval(as.name(sds_polish_mo[[1]]))
sdsPolishParMOT <- eval(as.name(sds_polish_mo[[2]]))
sdsPolishDecMO <- eval(as.name(sds_polish_mo[[3]]))
sdsPolishDecMOT <- eval(as.name(sds_polish_mo[[4]]))

sdsPolishParMV <- eval(as.name(sds_polish_mv[[1]]))
sdsPolishParMVT <- eval(as.name(sds_polish_mv[[2]]))
sdsPolishDecMV <- eval(as.name(sds_polish_mv[[3]]))
sdsPolishDecMVT <- eval(as.name(sds_polish_mv[[4]]))

sdsPolishSamp <- eval(as.name(sds_polish_samp[[1]]))
```

```r
# Test education sex relationship synthetic dataset for 1000 synthetic datasets
for (dataset_name in sds_names){
    dataset <- eval(as.name(dataset_name))
    
    male_voc_grm <- dplyr::filter(dataset$syn[[floor(runif(1, min=1, max=dataset$m))]], edu == 'VOCATIONAL/GRAMMAR' & sex == 'MALE')
    male_prim_noedu <- dplyr::filter(dataset$syn[[floor(runif(1, min=1, max=dataset$m))]], edu == 'PRIMARY/NO EDUCATION' & sex == 'MALE')
    male_sec <- dplyr::filter(dataset$syn[[floor(runif(1, min=1, max=dataset$m))]], edu == 'SECONDARY' & sex == 'MALE')
    male_post_sec_hgh <- dplyr::filter(dataset$syn[[floor(runif(1, min=1, max=dataset$m))]], edu == 'POST-SECONDARY OR HIGHER' & sex == 'MALE')
    female_voc_grm <- dplyr::filter(dataset$syn[[floor(runif(1, min=1, max=dataset$m))]], edu == 'VOCATIONAL/GRAMMAR' & sex == 'FEMALE')
    female_prim_noedu <- dplyr::filter(dataset$syn[[floor(runif(1, min=1, max=dataset$m))]], edu == 'PRIMARY/NO EDUCATION' & sex == 'FEMALE')
    female_sec <- dplyr::filter(dataset$syn[[floor(runif(1, min=1, max=dataset$m))]], edu == 'SECONDARY' & sex == 'FEMALE')
    female_post_sec_hgh <- dplyr::filter(dataset$syn[[floor(runif(1, min=1, max=dataset$m))]], edu == 'POST-SECONDARY OR HIGHER' & sex == 'FEMALE')
    
    total_count <- nrow(male_voc_grm) + nrow(male_prim_noedu) + nrow(male_sec) + nrow(male_post_sec_hgh) +
                    nrow(female_voc_grm) + nrow(female_prim_noedu) + nrow(female_sec) + nrow(female_post_sec_hgh)

    per_male_voc_grm <- (nrow(male_voc_grm)/total_count)*100
    per_male_prim_noedu <- (nrow(male_prim_noedu)/total_count)*100
    per_male_sec <- (nrow(male_sec)/total_count)*100
    per_male_post_sec_hgh <- (nrow(male_post_sec_hgh)/total_count)*100
    per_female_voc_grm <- (nrow(female_voc_grm)/total_count)*100
    per_female_prim_noedu <- (nrow(female_prim_noedu)/total_count)*100
    per_female_sec <- (nrow(female_sec)/total_count)*100
    per_female_post_sec_hgh <- (nrow(female_post_sec_hgh)/total_count)*100
    
    
    df_edu_sex[nrow(df_edu_sex) + 1,] <- c(dataset_name, per_male_voc_grm, per_male_prim_noedu, per_male_sec, per_male_post_sec_hgh,
                                          per_female_voc_grm, per_female_prim_noedu, per_female_sec, per_female_post_sec_hgh)
    
}
df_edu_sex$male_voc_grm <- as.numeric(df_edu_sex$male_voc_grm)
df_edu_sex$male_prim_noedu <- as.numeric(df_edu_sex$male_prim_noedu)
df_edu_sex$male_sec <- as.numeric(df_edu_sex$male_sec)
df_edu_sex$male_post_sec_hgh <- as.numeric(df_edu_sex$male_post_sec_hgh)
df_edu_sex$female_voc_grm <- as.numeric(df_edu_sex$female_voc_grm)
df_edu_sex$female_prim_noedu <- as.numeric(df_edu_sex$female_prim_noedu)
df_edu_sex$female_sec <- as.numeric(df_edu_sex$female_sec)
df_edu_sex$female_post_sec_hgh <- as.numeric(df_edu_sex$female_post_sec_hgh)
df_edu_sex
```
