```r
library('synthpop')
library('ggplot2')
library('ggrepel')
library('patchwork')
library(stringr)
load('../Synthetic_Datasets/SDS/Avila/Normal/Full/sdsAvilaM.RData')
load("../Synthetic_Datasets/SDS/Avila//Special//Full//sdsAvilaMS.RData")
load("../Synthetic_Datasets/SDS/Avila/Opposite/Full/sdsAvilaMO.RData")
load("../Synthetic_Datasets/SDS/Avila/Variant/Full/sdsAvilaMV.RData")
load("../Synthetic_Datasets/SDS/Avila/Random/sdsAvilaSamp.RData")
source('../Ex-Jobb//functions.R')
```

```r
# prepare original dataset
avilaODS <- read.csv("../Ex-Jobb/avila-tr.txt",header = FALSE)
colnames(avilaODS) <- c("Int_Dist","Up_Marg","Lo_Marg","Exploit","Row_Num","Mod_Ratio","Int_Spac","Weight","Peak_Num","MR/IS","Class")
```

```r
# prepare synthetic datasets
sds_avila_m <- paste0("sds","Avila",c("Par","Par","Dec","Dec"),"M",c("","T"))
sds_avila_cat_cac <- paste0("sds","AvilaCatM",c("P","P","C","C"),"",c("","T"))
sds_avila_mo <- paste0("sds","Avila",c("Par","Par","Dec","Dec"),"MO",c("","T"))
sds_avila_mv <- paste0("sds","Avila",c("Par","Par","Dec","Dec"),"MV",c("","T"))
sds_avila_samp <- paste0("sds","AvilaSamp",c(""),"",c(""))
sds_names <- c(sds_avila_m, sds_avila_cat_cac, sds_avila_mo, sds_avila_mv, sds_avila_samp)
sds_names

sds_names <- str_sort(sds_names)
sds_names

sdsAvilaParM <- eval(as.name(sds_avila_m[[1]]))
sdsAvilaParMT <- eval(as.name(sds_avila_m[[2]]))
sdsAvilaDecM <- eval(as.name(sds_avila_m[[3]]))
sdsAvilaDecMT <- eval(as.name(sds_avila_m[[4]]))

sdsAvilaCatMP <- eval(as.name(sds_avila_cat_cac[[1]]))
sdsAvilaCatMPT <- eval(as.name(sds_avila_cat_cac[[2]]))
sdsAvilaCatMC <- eval(as.name(sds_avila_cat_cac[[3]]))
sdsAvilaCatMCT <- eval(as.name(sds_avila_cat_cac[[4]]))

sdsAvilaParMO <- eval(as.name(sds_avila_mo[[1]]))
sdsAvilaParMOT <- eval(as.name(sds_avila_mo[[2]]))
sdsAvilaDecMO <- eval(as.name(sds_avila_mo[[3]]))
sdsAvilaDecMOT <- eval(as.name(sds_avila_mo[[4]]))

sdsAvilaParMV <- eval(as.name(sds_avila_mv[[1]]))
sdsAvilaParMVT <- eval(as.name(sds_avila_mv[[2]]))
sdsAvilaDecMV <- eval(as.name(sds_avila_mv[[3]]))
sdsAvilaDecMVT <- eval(as.name(sds_avila_mv[[4]]))

sdsAvilaamp <- eval(as.name(sds_avila_samp[[1]]))
```

```r
# Test univariate distribution of MR/IS Variable
df <-data.frame(Variant= "Original", Values = avilaODS$`MR/IS`)
corr_names <- c("CC", "CCT", "CP", "CPT", "D", "DO", "DOT", "DT", "DV", "DVT", "P", "PO", "POT", "PT", "PV", "PVT", "S")
for (i in c(1:length(sds_names))){
    act_avila <- eval(as.name(sds_names[[i]]))
    tdf <- data.frame(Variant= corr_names[[i]], Values = act_avila$syn[[2]]$`MR/IS`)
    df <- rbind(df, tdf)
}
```

```r
# Plot univariate distribution of MR/IS Variable
library(tidyverse)
library(ggjoy)


plot_msis <- ggplot(df, aes(x=Values, y=Variant))+
    geom_joy(scale = 2, alpha=0.5) +
    scale_y_discrete(expand=c(0.01, 0)) +
    scale_x_continuous(expand=c(0.01, 0)) +
    theme_joy()+
    labs(title = "Distribution of MR/IS Variable", 
         y = "")
plot_msis
```

```r
# Test univariate distribution of Exploit Variable

df <-data.frame(Variant= "Original", Values = avilaODS$Exploit)
corr_names <- c("CC", "CCT", "CP", "CPT", "D", "DO", "DOT", "DT", "DV", "DVT", "P", "PO", "POT", "PT", "PV", "PVT", "S")
for (i in c(1:length(sds_names))){
    act_avila <- eval(as.name(sds_names[[i]]))
    tdf <- data.frame(Variant= corr_names[[i]], Values = act_avila$syn[[2]]$Exploit)
    df <- rbind(df, tdf)
}
```

```r
# Plot univariate distribution of Exploit Variable

plot_exploit <- ggplot(df, aes(x=Values, y=Variant))+
    geom_joy(scale = 2, alpha=0.5) +
    scale_y_discrete(expand=c(0.01, 0)) +
    scale_x_continuous(expand=c(0.01, 0)) +
    theme_joy()+
    labs(title = "Distribution of Exploit Variable", 
         y = "")
plot_exploit
```

```r
# Test Units of Authors for synthetic dataset
df_class_a = data.frame(matrix(ncol = 17, nrow = 0))
df_class_f = data.frame(matrix(ncol = 17, nrow = 0))
df_class_h = data.frame(matrix(ncol = 17, nrow = 0))
df_class_e = data.frame(matrix(ncol = 17, nrow = 0))
df_class_i = data.frame(matrix(ncol = 17, nrow = 0))
df_class_y = data.frame(matrix(ncol = 17, nrow = 0))
df_class_d = data.frame(matrix(ncol = 17, nrow = 0))
df_class_x = data.frame(matrix(ncol = 17, nrow = 0))
df_class_g = data.frame(matrix(ncol = 17, nrow = 0))
df_class_w = data.frame(matrix(ncol = 17, nrow = 0))
df_class_c = data.frame(matrix(ncol = 17, nrow = 0))
df_class_b = data.frame(matrix(ncol = 17, nrow = 0))
colnames(df_class_a) <- sds_names
colnames(df_class_f) <- sds_names
colnames(df_class_h) <- sds_names
colnames(df_class_e) <- sds_names
colnames(df_class_i) <- sds_names
colnames(df_class_y) <- sds_names
colnames(df_class_d) <- sds_names
colnames(df_class_x) <- sds_names
colnames(df_class_g) <- sds_names
colnames(df_class_w) <- sds_names
colnames(df_class_c) <- sds_names
colnames(df_class_b) <- sds_names
#sds_names_test <- paste0('sdsAdultCatMC')
for (dataset_name in sds_names){
    dataset <- eval(as.name(dataset_name))
    for (i in 1:dataset$m){                                                            
        class_a <- dplyr::filter(dataset$syn[[i]], Class == 'A')
        class_f <- dplyr::filter(dataset$syn[[i]], Class == 'F')
        class_h <- dplyr::filter(dataset$syn[[i]], Class == 'H')
        class_e <- dplyr::filter(dataset$syn[[i]], Class == 'E')
        class_i <- dplyr::filter(dataset$syn[[i]], Class == 'I')
        class_y <- dplyr::filter(dataset$syn[[i]], Class == 'Y')
        class_d <- dplyr::filter(dataset$syn[[i]], Class == 'D')
        class_x <- dplyr::filter(dataset$syn[[i]], Class == 'X')
        class_g <- dplyr::filter(dataset$syn[[i]], Class == 'G')
        class_w <- dplyr::filter(dataset$syn[[i]], Class == 'W')
        class_c <- dplyr::filter(dataset$syn[[i]], Class == 'C')
        class_b <- dplyr::filter(dataset$syn[[i]], Class == 'B')

        df_class_a[i, dataset_name] <- c(nrow(class_a))
        df_class_f[i, dataset_name] <- c(nrow(class_f))
        df_class_h[i, dataset_name] <- c(nrow(class_h))
        df_class_e[i, dataset_name] <- c(nrow(class_e))
        df_class_i[i, dataset_name] <- c(nrow(class_i))
        df_class_y[i, dataset_name] <- c(nrow(class_y))
        df_class_d[i, dataset_name] <- c(nrow(class_d))
        df_class_x[i, dataset_name] <- c(nrow(class_x))
        df_class_g[i, dataset_name] <- c(nrow(class_g))
        df_class_w[i, dataset_name] <- c(nrow(class_w))
        df_class_c[i, dataset_name] <- c(nrow(class_c))
        df_class_b[i, dataset_name] <- c(nrow(class_b))
        }
    
}
```
