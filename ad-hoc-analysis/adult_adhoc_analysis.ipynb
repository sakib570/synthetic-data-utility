```r
# prepare original dataset
adultODS <- read.obs("../Ex-Jobb/adultdata.csv")
adultODSM <- adultODS

for(i in c(1:ncol(adultODS))){
  remAD <- is.na(adultODSM[,i])
  adultODSM <- adultODSM[!remAD,]
}

adultODSM <- adultODSM[,-c(3,5)]
#eduLev <- adultODSM$education[c(14,4,5,6,7,3,15,11)]
eduLev <- c('Assoc-voc', 'Some-college', 'Bachelors', 'HS-grad', 'Masters', 'Prof-school', 'Assoc-acdm', '11th', 
            'Doctorate', '12th', '5th-6th', '10th', '9th', '1st-4th', '7th-8th', 'Preschool')
adultODSM$education <- factor(adultODSM$education,eduLev)
adultODSM$relationship <- factor(adultODSM$relationship,
                                 c("Not-in-family","Other-relative","Own-child","Unmarried","Wife","Husband"))

Europe <- c("England","France","Germany","Greece","Holand-Netherlands",
            "Hungary","Ireland","Italy","Poland","Portugal","Scotland",
            "Yugoslavia")
Asia <- c("Cambodia","China","Hong","India","Iran","Japan","Laos",
          "Philippines","South","Taiwan","Thailand","Vietnam")
OtherUS <- c("Outlying-US(Guam-USVI-etc)","Puerto-Rico")
SouthAmerica <- c("Columbia","Cuba","Dominican-Republic","Ecuador",
                  "El-Salvador","Guatemala","Haiti","Honduras",
                  "Jamaica","Nicaragua","Peru","Trinadad&Tobago")

adultODSM2 <- adultODSM
adultODSM2$native.country <- as.character(adultODSM2$native.country)
for(cs in Europe){
  adultODSM2$native.country[adultODSM2$native.country == cs] <- "Europe"
}
for(cs in Asia){
  adultODSM2$native.country[adultODSM2$native.country == cs] <- "Asia"
}
for(cs in OtherUS){
  adultODSM2$native.country[adultODSM2$native.country == cs] <- "OtherUS"
}
for(cs in SouthAmerica){
  adultODSM2$native.country[adultODSM2$native.country == cs] <- "South-America"
}

adultODSM2$native.country <- factor(adultODSM2$native.country,
                                    c("United-States","OtherUS","Canada","Mexico",
                                      "South-America","Europe","Asia"))

adultODS10K <- adultODSM2[1:10000,]
adultODS10_to_15 <- adultODSM2[10001:15000,]
#write.csv(adultODS10K,"adultODS10K.csv", row.names = FALSE)
#write.csv(adultODS10_to_15,"../adultODS10K_to_15K.csv", row.names = FALSE)
```

```r
# Test income, gender & native vountry relation for single dataset
df_income_sex_us = data.frame(matrix(ncol = 5, nrow = 0))
colnames(df_income_sex_us) <- c("name", "male_below_50", "male_above_50", "female_below_50", "female_above_50")

org_male_below_50 <- dplyr::filter(adultODS10K, income == '<=50K' & sex == 'Male' & native.country == "United-States")
org_male_above_50 <- dplyr::filter(adultODS10K, income == '>50K' & sex == 'Male' & native.country == "United-States")
org_female_below_50 <- dplyr::filter(adultODS10K, income == '<=50K' & sex == 'Female' & native.country == "United-States")
org_female_above_50 <- dplyr::filter(adultODS10K, income == '>50K' & sex == 'Female' & native.country == "United-States")
org_total_count <- nrow(org_male_below_50) + nrow(org_male_above_50) + nrow(org_female_below_50) + nrow(org_female_above_50)
org_per_male_below_50 <- (nrow(org_male_below_50)/org_total_count)*100
org_per_male_above_50 <- (nrow(org_male_above_50)/org_total_count)*100
org_per_female_below_50 <- (nrow(org_female_below_50)/org_total_count)*100
org_per_female_above_50 <- (nrow(org_female_above_50)/org_total_count)*100
df_income_sex_us[nrow(df_income_sex_us) + 1,] <- c("Original", org_per_male_below_50, org_per_male_above_50, org_per_female_below_50, org_per_female_above_50)

for (dataset_name in sds_names){
    dataset <- eval(as.name(dataset_name))
    male_below_50 <- dplyr::filter(dataset$syn[[floor(runif(1, min=1, max=dataset$m))]], income == '<=50K' & sex == 'Male' & native.country == "United-States")
    male_above_50 <- dplyr::filter(dataset$syn[[floor(runif(1, min=1, max=dataset$m))]], income == '>50K' & sex == 'Male' & native.country == "United-States")
    female_below_50 <- dplyr::filter(dataset$syn[[floor(runif(1, min=1, max=dataset$m))]], income == '<=50K' & sex == 'Female' & native.country == "United-States")
    female_above_50 <- dplyr::filter(dataset$syn[[floor(runif(1, min=1, max=dataset$m))]], income == '>50K' & sex == 'Female' & native.country == "United-States")
    total_count <- nrow(male_below_50) + nrow(male_above_50) + nrow(female_below_50) + nrow(female_above_50)
    per_male_below_50 <- (nrow(male_below_50)/total_count)*100
    per_male_above_50 <- (nrow(male_above_50)/total_count)*100
    per_female_below_50 <- (nrow(female_below_50)/total_count)*100
    per_female_above_50 <- (nrow(female_above_50)/total_count)*100
    df_income_sex_us[nrow(df_income_sex_us) + 1,] <- c(dataset_name, per_male_below_50, per_male_above_50, per_female_below_50, per_female_above_50)
    
}
df_income_sex_us$male_below_50 <- as.numeric(df_income_sex_us$male_below_50)
df_income_sex_us$male_above_50 <- as.numeric(df_income_sex_us$male_above_50)
df_income_sex_us$female_below_50 <- as.numeric(df_income_sex_us$female_below_50)
df_income_sex_us$female_above_50 <- as.numeric(df_income_sex_us$female_above_50)
df_income_sex_us
```
